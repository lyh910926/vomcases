# Weatherdata


```python
import os
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime, timedelta, date
from IPython.display import display
```

## Timeseries

```python
data_adelaide = np.genfromtxt("../../data/VOM_input/AdelaideRiver/dailyweather.prn",names=True  )
data_daly = np.genfromtxt("../../data/VOM_input/DalyUncleared/dailyweather.prn",names=True  )
data_dry = np.genfromtxt("../../data/VOM_input/DryRiver/dailyweather.prn",names=True  )
data_howard = np.genfromtxt("../../data/VOM_input/HowardSprings/dailyweather.prn",names=True  )
data_sturt = np.genfromtxt("../../data/VOM_input/SturtPlains/dailyweather.prn",names=True  )
data_litchfield = np.genfromtxt("../../data/VOM_input/Litchfield/dailyweather.prn",names=True  )
```

```python
time = pd.date_range(str(int(data_adelaide["Day"][0])) + "-" +str(int(data_adelaide["Month"][0])) + "-" + str(int(data_adelaide["Year"][0])), periods=len(data_adelaide["Dcum"]), freq='D')   
```

```python
howard_pd = pd.DataFrame(data_howard, index=time)
adelaide_pd = pd.DataFrame(data_adelaide, index=time)
litch_pd = pd.DataFrame(data_litchfield, index=time)
daly_pd = pd.DataFrame(data_daly, index=time)
dry_pd = pd.DataFrame(data_dry, index=time)
sturt_pd = pd.DataFrame(data_sturt, index=time)
```

```python
fig, ax   = plt.subplots(nrows=6, ncols=1, figsize=(24, 30), sharex=True, sharey=True)   
ax[0].plot(howard_pd["Rain"])
ax[0].set_ylabel("Precipitation (mm/d)", size=16)
ax[1].plot(adelaide_pd["Rain"])
ax[1].set_ylabel("Precipitation (mm/d)", size=16)

ax[2].plot(litch_pd["Rain"])
ax[2].set_ylabel("Precipitation (mm/d)", size=16)

ax[3].plot(daly_pd["Rain"])
ax[3].set_ylabel("Precipitation (mm/d)", size=16)

ax[4].plot(dry_pd["Rain"])
ax[4].set_ylabel("Precipitation (mm/d)", size=16)

ax[5].plot(sturt_pd["Rain"])
ax[5].set_ylabel("Precipitation (mm/d)", size=16)

plt.show()
```

```python
fig, ax   = plt.subplots(nrows=6, ncols=1, figsize=(24, 30), sharex=True, sharey=True)   
ax[0].plot(howard_pd["TMax"])
ax[0].plot(howard_pd["TMin"])
ax[0].set_ylabel(r"Temperature ($^o$C)", size=16)

ax[1].plot(adelaide_pd["TMax"])
ax[1].plot(adelaide_pd["TMin"])
ax[1].set_ylabel(r"Temperature ($^o$C)", size=16)

ax[2].plot(litch_pd["TMax"])
ax[2].plot(litch_pd["TMin"])
ax[2].set_ylabel(r"Temperature ($^o$C)", size=16)

ax[3].plot(daly_pd["TMax"])
ax[3].plot(daly_pd["TMin"])
ax[3].set_ylabel(r"Temperature ($^o$C)", size=16)

ax[4].plot(dry_pd["TMax"])
ax[4].plot(dry_pd["TMin"])
ax[4].set_ylabel(r"Temperature ($^o$C)", size=16)

ax[5].plot(sturt_pd["TMax"])
ax[5].plot(sturt_pd["TMin"])
ax[5].set_ylabel(r"Temperature ($^o$C)", size=16)

plt.show()
```

```python
fig, ax   = plt.subplots(nrows=6, ncols=1, figsize=(24, 40), sharex=True, sharey=True)   
ax[0].plot(howard_pd["Radn"])
ax[0].set_ylabel(r"Incoming shortwave radiation (MJ/m$^2$)", size=16)

ax[1].plot(adelaide_pd["Radn"])
ax[1].set_ylabel(r"Incoming shortwave radiation (MJ/m$^2$)", size=16)

ax[2].plot(litch_pd["Radn"])
ax[2].set_ylabel(r"Incoming shortwave radiation (MJ/m$^2$)", size=16)

ax[3].plot(daly_pd["Radn"])
ax[3].set_ylabel(r"Incoming shortwave radiation (MJ/m$^2$)", size=16)

ax[4].plot(dry_pd["Radn"])
ax[4].set_ylabel(r"Incoming shortwave radiation (MJ/m$^2$)", size=16)

ax[5].plot(sturt_pd["Radn"])
ax[5].set_ylabel(r"Incoming shortwave radiation (MJ/m$^2$)", size=16)

plt.show()
```

```python
fig, ax   = plt.subplots(nrows=6, ncols=1, figsize=(24, 40), sharex=True, sharey=True)   
ax[0].plot(howard_pd["VP"])
ax[0].set_ylabel(r"Vapour pressure (hPa)", size=16)

ax[1].plot(adelaide_pd["VP"])
ax[1].set_ylabel(r"Vapour pressure (hPa)", size=16)

ax[2].plot(litch_pd["VP"])
ax[2].set_ylabel(r"Vapour pressure (hPa)", size=16)

ax[3].plot(daly_pd["VP"])
ax[3].set_ylabel(r"Vapour pressure (hPa)", size=16)

ax[4].plot(dry_pd["VP"])
ax[4].set_ylabel(r"Vapour pressure (hPa)", size=16)

ax[5].plot(sturt_pd["VP"])
ax[5].set_ylabel(r"Vapour pressure (hPa)", size=16)

plt.show()
```

```python
fig, ax   = plt.subplots(nrows=6, ncols=1, figsize=(24, 40), sharex=True, sharey=True)   
ax[0].plot(howard_pd["Pres"])
ax[0].set_ylabel(r"Atmospheric pressure (hPa)", size=16)

ax[1].plot(adelaide_pd["Pres"])
ax[1].set_ylabel(r"Atmospheric pressure (hPa)", size=16)

ax[2].plot(litch_pd["Pres"])
ax[2].set_ylabel(r"Atmospheric pressure (hPa)", size=16)

ax[3].plot(daly_pd["Pres"])
ax[3].set_ylabel(r"Atmospheric pressure (hPa)", size=16)

ax[4].plot(dry_pd["Pres"])
ax[4].set_ylabel(r"Atmospheric pressure (hPa)", size=16)

ax[5].plot(sturt_pd["Pres"])
ax[5].set_ylabel(r"Atmospheric pressure (hPa)", size=16)

plt.show()
```

## Mean Annual Values

```python
MAP_hs = np.mean(howard_pd["Rain"].resample("A").sum())
MAP_ade = np.mean(adelaide_pd["Rain"].resample("A").sum())
MAP_litch = np.mean(litch_pd["Rain"].resample("A").sum())
MAP_daly = np.mean(daly_pd["Rain"].resample("A").sum())
MAP_dry = np.mean(dry_pd["Rain"].resample("A").sum())
MAP_sturt = np.mean(sturt_pd["Rain"].resample("A").sum())
```

```python
print("HowardSprings | " + str(MAP_hs) + " mm/year" )
print("AdelaideRiver | " + str(MAP_ade) + " mm/year" )
print("Litchfield    | " + str(MAP_litch) + " mm/year" )
print("DalyUncleared | " + str(MAP_daly) + " mm/year" )
print("DryRiver      | " + str(MAP_dry) + " mm/year" )
print("SturtPlains   | " + str(MAP_sturt) + " mm/year" )
```
