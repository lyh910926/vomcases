## Timeseries

### Precipitation

```python
data_adelaide = np.genfromtxt("../../data/VOM_input/AdelaideRiver/dailyweather.prn",names=True  )
data_daly = np.genfromtxt("../../data/VOM_input/DalyUncleared/dailyweather.prn",names=True  )
data_dry = np.genfromtxt("../../data/VOM_input/DryRiver/dailyweather.prn",names=True  )
data_howard = np.genfromtxt("../../data/VOM_input/HowardSprings/dailyweather.prn",names=True  )
data_sturt = np.genfromtxt("../../data/VOM_input/SturtPlains/dailyweather.prn",names=True  )
data_litchfield = np.genfromtxt("../../data/VOM_input/Litchfield/dailyweather.prn",names=True  )
```

```python
time = pd.date_range(str(int(data_adelaide["Day"][0])) + "-" +str(int(data_adelaide["Month"][0])) + "-" + str(int(data_adelaide["Year"][0])), periods=len(data_adelaide["Dcum"]), freq='D')   
```

```python
howard_pd = pd.DataFrame(data_howard, index=time)
adelaide_pd = pd.DataFrame(data_adelaide, index=time)
litch_pd = pd.DataFrame(data_litchfield, index=time)
daly_pd = pd.DataFrame(data_daly, index=time)
dry_pd = pd.DataFrame(data_dry, index=time)
sturt_pd = pd.DataFrame(data_sturt, index=time)
```

```python
fig, ax   = plt.subplots(nrows=6, ncols=1, figsize=(18, 20), sharex=True, sharey=True)   
ax[0].plot(howard_pd["Rain"])
ax[0].set_ylabel("Precipitation (mm/d)", size=20)
ax[0].text(0.01, 0.9, "a)", ha='left', va='center', transform=ax[0].transAxes, fontsize=20)

ax[1].plot(litch_pd["Rain"])
ax[1].set_ylabel("Precipitation (mm/d)", size=20)
ax[1].text(0.01, 0.9, "b)", ha='left', va='center', transform=ax[1].transAxes, fontsize=20)

ax[2].plot(adelaide_pd["Rain"])
ax[2].set_ylabel("Precipitation (mm/d)", size=20)
ax[2].text(0.01, 0.9, "c)", ha='left', va='center', transform=ax[2].transAxes, fontsize=20)

ax[3].plot(daly_pd["Rain"])
ax[3].set_ylabel("Precipitation (mm/d)", size=20)
ax[3].text(0.01, 0.9, "d)", ha='left', va='center', transform=ax[3].transAxes, fontsize=20)

ax[4].plot(dry_pd["Rain"])
ax[4].set_ylabel("Precipitation (mm/d)", size=16)
ax[4].set_ylabel("Precipitation (mm/d)", size=20)
ax[4].text(0.01, 0.9, "e)", ha='left', va='center', transform=ax[4].transAxes, fontsize=20)

ax[5].plot(sturt_pd["Rain"])
ax[5].set_ylabel("Precipitation (mm/d)", size=16)
ax[5].set_ylabel("Precipitation (mm/d)", size=20)
ax[5].text(0.01, 0.9, "f)", ha='left', va='center', transform=ax[5].transAxes, fontsize=20)

plt.tight_layout()
plt.show()
```

## Mean Annual Values

```python
MAP_hs = np.mean(howard_pd["Rain"].resample("A").sum())
MAP_ade = np.mean(adelaide_pd["Rain"].resample("A").sum())
MAP_litch = np.mean(litch_pd["Rain"].resample("A").sum())
MAP_daly = np.mean(daly_pd["Rain"].resample("A").sum())
MAP_dry = np.mean(dry_pd["Rain"].resample("A").sum())
MAP_sturt = np.mean(sturt_pd["Rain"].resample("A").sum())
```

```python
MeanMaxT_hs = np.mean(howard_pd["TMax"].resample("A").max())
MeanMaxT_ade = np.mean(adelaide_pd["TMax"].resample("A").max())
MeanMaxT_litch = np.mean(litch_pd["TMax"].resample("A").max())
MeanMaxT_daly = np.mean(daly_pd["TMax"].resample("A").max())
MeanMaxT_dry = np.mean(dry_pd["TMax"].resample("A").max())
MeanMaxT_sturt = np.mean(sturt_pd["TMax"].resample("A").max())
```

```python
MeanMinT_hs = np.mean(howard_pd["TMin"].resample("A").max())
MeanMinT_ade = np.mean(adelaide_pd["TMin"].resample("A").max())
MeanMinT_litch = np.mean(litch_pd["TMin"].resample("A").max())
MeanMinT_daly = np.mean(daly_pd["TMin"].resample("A").max())
MeanMinT_dry = np.mean(dry_pd["TMin"].resample("A").max())
MeanMinT_sturt = np.mean(sturt_pd["TMin"].resample("A").max())
```

```python
data = [[ MAP_hs, MeanMinT_hs,  MeanMaxT_hs],
        [ MAP_litch, MeanMinT_litch,  MeanMaxT_litch],
        [ MAP_ade,  MeanMinT_ade, MeanMaxT_ade],
        [ MAP_daly,  MeanMinT_daly, MeanMaxT_daly],
        [MAP_dry, MeanMinT_dry, MeanMaxT_dry],
        [MAP_sturt, MeanMinT_sturt, MeanMaxT_sturt], ]

rows = ('Howard Springs', 'Litchfield', 'Adelaide River', 'Daly Uncleared', 'Dry River', 'Sturt Plains')
columns = ( "prec. (mm/year)", "max.temp $^o$C",  "min.temp $^o$C"    )

n_rows = len(data)

cell_text = []
for row in range(n_rows):
    cell_text.append(['%1.1f' % x  for x in data[row]])


fig.patch.set_visible(False)
plt.axis('off')
plt.axis('tight')

# Add a table at the bottom of the axes
the_table = plt.table(cellText=cell_text,
                      rowLabels=rows,
                      colLabels=columns,
                      loc='center')

the_table.auto_set_font_size(False)
the_table.set_fontsize(18)
the_table.scale(2, 2)

plt.show()
```

**Table S7.1** Mean annual values of precipitation, daily maximum and daily minimum temperatures for the different sites.
